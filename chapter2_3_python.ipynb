```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the data file
url = 'https://raw.githubusercontent.com/ogut77/DataScience/refs/heads/main/PurchaseByCredit40.csv'
card = pd.read_csv(url)


```

```python
# Display the data
print(card)
```

```python
# Create a bar plot
plt.figure(figsize=(6, 6))  # Set figure size for better visibility
card['Purchase'].value_counts().plot(kind='bar', title='Bar Graph', ylabel='frequency', ylim=(0, 30))
plt.show()
```

```python
freq = card['Purchase'].value_counts()
# Plot a pie chart
print(freq)
plt.pie(freq, labels=freq.index)
plt.title("Pie Chart")
plt.show()
```

```python


import pandas as pd

# Load Iris dataset from UCI Machine Learning Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width', 'Species']
iris= pd.read_csv(url, header=None, names=columns)
#Display the data
iris

```

```python

import matplotlib.pyplot as plt
# Create a histogram for 'sepal length (cm)'
plt.figure(figsize=(8, 6))
plt.hist(iris['Sepal.Length'], bins=10, edgecolor='black')  # Adjust bins as needed
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python

import matplotlib.pyplot as plt

# Assuming iris_df is already defined as in the previous code

plt.figure(figsize=(8, 6))
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(iris, vars=['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width'])

# Show the plot
plt.show()
```

```python
import seaborn as sns
sns.pairplot(iris, vars=['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width'], hue='Species')
# Show the plot
plt.show()
```

```python
import plotly.express as px

# Assuming 'iris' DataFrame is already loaded as in the previous code

# Create parallel coordinates plot using Plotly
species_map = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
iris['Species_numeric'] = iris['Species'].map(species_map)
fig = px.parallel_coordinates(iris,
                             dimensions=['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width'],
                             color="Species_numeric",
                             color_continuous_scale=px.colors.qualitative.Plotly)  # Use a qualitative colorscale for categorical data
fig.show()
```
